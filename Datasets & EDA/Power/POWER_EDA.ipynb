```
import pandas as pd
import numpy as np
```

```
dtypes = {'Global_active_power': np.float64, 'Global_reactive_power': np.float64, 'Voltage':np.float64, 'Global_intensity':np.float64, 'Sub_metering_1':np.float64, 
'Sub_metering_2':np.float64}
```

```
data = pd.read_csv('household_power_consumption.zip', sep=';', dtype=dtypes, na_values=['?'])
```

```
data.dtypes
```

```
data.head()
```

```
data['Global_active_power'].plot()
```

```
pd.to_datetime(data['Date'] + ':' + data['Time'], format='%d/%m/%Y:%H:%M:%S')
```

```
data['Datetime'] = pd.to_datetime(data['Date'] + ':' + data['Time'], format='%d/%m/%Y:%H:%M:%S')
```

```
data.head()
```

```
data.interpolate()
```

```
processed_data = data.interpolate()
```
